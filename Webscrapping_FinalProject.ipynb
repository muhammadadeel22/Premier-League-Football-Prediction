```python
years = list(range(2023, 2021, -1))
all_matches = []
```

```python
standings_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
```

```python
import time
import requests
import pandas as pd
import io
from bs4 import BeautifulSoup

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

for year in years:
    response = requests.get(standings_url, headers=headers)
    if response.status_code == 429:
        print("Rate limited. Retrying after delay...")
        time.sleep(60)
        continue
    elif response.status_code != 200:
        print(f"Failed to fetch {standings_url}: {response.status_code}")
        break

    soup = BeautifulSoup(response.text, 'html.parser')
    tables = soup.select('table.stats_table')
    if not tables:
        print(f"No standings table found for {standings_url}")
        break

    standings_table = tables[0]
    links = [l.get("href") for l in standings_table.find_all('a')]
    links = [l for l in links if '/squads/' in l]
    team_urls = [f"https://fbref.com{l}" for l in links]

    previous_season = soup.select("a.prev")
    if not previous_season:
        print(f"No link to previous season found for {standings_url}")
        break
    standings_url = f"https://fbref.com{previous_season[0].get('href')}"

    for team_url in team_urls:
        time.sleep(5)  # Respect rate limits
        team_name = team_url.split("/")[-1].replace("-Stats", "").replace("-", " ")
        data = requests.get(team_url, headers=headers)

        try:
            # Parse "Scores & Fixtures"
            matches = pd.read_html(io.StringIO(data.text), match="Scores & Fixtures")[0]
        except ValueError:
            print(f"Scores & Fixtures table not found for {team_name}")
            continue

        soup = BeautifulSoup(data.text, 'html.parser')
        links = [l.get("href") for l in soup.find_all('a')]
        links = [l for l in links if l and 'all_comps/shooting/' in l]

        if not links:
            print(f"Shooting stats link not found for {team_name}")
            continue

        shooting_url = f"https://fbref.com{links[0]}"
        data = requests.get(shooting_url, headers=headers)

        try:
            # Parse "Shooting" stats
            shooting = pd.read_html(io.StringIO(data.text), match="Shooting")[0]
        except ValueError:
            print(f"Shooting table not found for {team_name}")
            continue

        shooting.columns = shooting.columns.droplevel()
        try:
            team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
        except ValueError as e:
            print(f"Error merging data for {team_name}: {e}")
            continue

        team_data = team_data[team_data["Comp"] == "Premier League"]
        #team_data["Season"] = year
        team_data["Team"] = team_name
        all_matches.append(team_data)
        time.sleep(5)
```

```python
len(all_matches)
```

```python
match_df = pd.concat(all_matches)
```

```python
match_df.columns = [c.lower() for c in match_df.columns]
```
